```python
import requests
from bs4 import BeautifulSoup as bs
import os
import sys
import time
import random
from io import BytesIO
from PIL import Image
import re

import json

s=requests.Session()
s.keep_alive = False

headers ={
    'accept-encoding': 'gzip, deflate, br',
    'accept-language': 'zh-CN,zh;q=0.9',
    'upgrade-insecure-requests': '1',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.114 Safari/537.36'
}




if __name__=="__main__":
    print("begin to crawl Auto pics!")
    url = "https://www.auto.com/cars-for-sale"
    path = 'D:/datasets/Carimages/'
    ret =s.get(url,headers=headers)
    
    if ret.status_code == 200:
        print("get function successful!<<<<<")
        soup = bs(ret.text,"lxml")
        
        # 直接找到href的地址，这里不循坏了
        a = soup.select("ep-accordion ep-link-pack")[2]
        url =[]
        for i in a:
            b= str(i)
            try:
                url.append(b.split(r"<a href=")[1].split(r">")[0])
            except:
                pass
        foldername =""
        count = 0
        for urls in url:
            count+=1
            if count>22:   
                newurl = "https://www.auto.com"+eval(urls)
                foldername = eval(urls).split("/")[-1]
                foldpath = path+foldername
                if os.path.exists(foldpath)==False:
                    os.makedirs(foldpath)
                print("folder:",foldername)
                print('newurl',newurl)
                rets = s.get(newurl,headers=headers)
                if rets.status_code ==200:
                    #print("access sub class models<<<<")
                    soups = bs(rets.text,"lxml")
                    #print(soup)
                    #根据属性进行搜索，refer to "https://www.jb51.net/article/201558.htm"
                    # 这里第二个元素包含json 需要的href 源
                    retss= soups.find_all(attrs={"type":"application/ld+json"})[1]  
                    #转化为string类型
                    retss = str(retss)
                    reg = r'<script type="application/ld+json">'
                    a=retss.split(reg)[1].split('</script>')[0]
                    a= json.loads(a)
                    for i in a:
                        newurl=i['offers']['url']
                        subclass = s.get(newurl,headers=headers)
                        if subclass.status_code ==200:
                            print("acces into subclass,ready to download!!<<<<<")
                            soupp = bs(subclass.text,"html.parser")
                            for srcs in soupp.select("ep-image img"):
                                print("srcs:",srcs["data-src"])
                                finalsrc = srcs["data-src"]
                                if "png" in finalsrc[-4:]:  
                                    picname = finalsrc.split("/")[-1]
                                elif "jpg" in finalsrc[-4:]:
                                    picname = finalsrc.split("/")[-1]
                                print('picname:',picname)
                                abspicname = os.path.join(foldpath,picname)
                                print('abspicname:',abspicname)

                                ht=s.get(finalsrc)
                                if ht.status_code == 200: 
                                    file = BytesIO(ht.content)
                                    try: 
                                        img = Image.open(file)                  
                                        img.save(abspicname)
                                    except:
                                        print("open file failure!, continue")
                                        continue

    else:
        print("return failure code")
    
```
