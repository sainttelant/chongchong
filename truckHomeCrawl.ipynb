```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup as bs
from PIL import Image
import datetime
from urllib import parse
import re
import random
import time
from io import BytesIO
import json
import os

s=requests.Session()
headers={
            'accept':'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
            'accept-encoding':'gzip, deflate, br',
            'accept-language':'en-US,en;q=0.8',
            'referer':'https://product.360che.com/BrandList.html',
            'upgrade-insecure-requests':'1',
            'user-agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'
        }

url='https://product.360che.com/BrandList-1.html'
subclass_list=[]
suburl_list=[]
os.makedirs("truckLogs")


def title(url):
    classlist=[]
    html=s.get(url,headers=headers)
    soup=bs(html.text,'lxml')
    classoftruck=soup.select(".sel-act")[0].get_text()
    subclass=soup.select("div.xll_center2_a1_y1 a")
    for i in subclass:
        classlist.append(i.get_text())
    return classoftruck,classlist
def crawlPic(url):
    html=s.get(url,headers=headers)
    soup=bs(html.text,"lxml")
    picoverall=soup.select("div.xll_center2_a1_z dl dt a img")
    for i in picoverall:
        picurl=i['src']
        try:
            if "/" in i['title']:
                inew=i['title'].replace("/","_")
                picname=inew+i["src"].split("https://img4.kcimg.cn/imgser/")[1].split("!")[0]
            else: 
                picname=i['title']+i["src"].split("https://img4.kcimg.cn/imgser/")[1].split("!")[0]
            piccontent=s.get(picurl,headers=headers)
            if piccontent.status_code==200:
                print(i["title"])
                print(i["src"])
                f=open('truckLogs/'+picname,"wb")
                f.write(piccontent.content)
                f.close()
        except IndexError:
            print("ingore no pictitle's problem")
        else:
            print("it should be downloaded!")
            
        

def suburl():
    html=s.get(url,headers=headers).content
    soup=bs(html,'html.parser')
    suburls=soup.select("div.lbpd-box a")
    for i in suburls:
        newurl=url.split("BrandList-1.html")[0]+i['href']
        suburl_list.append(newurl)
    return suburl_list


if __name__=="__main__":
    print("<<<<<<<<It is begin to crawl car classifier for marketing depart>>>>>>")
    data_d=dict()
    suburls_L=suburl()
    for url in suburls_L:
        Bigclass,subclass=title(url)
        
        crawlPic(url)
        data_d[Bigclass]=subclass
    print("all done")
```
