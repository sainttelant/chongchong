```python
import requests
from bs4 import BeautifulSoup as bs
from multiprocessing import Pool
import os,re
import time

s=requests.Session()
s.keep_alive = False
headers={'Connection':'keep-alive',
'Host':'hm.baidu.com',
'Referer':'https://www.219ii.com/htm/piclist6/',
'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}
pageNum=15
img_url=[]
name=[]
tupurl=[]

def get_img_url():
    for i in range(1,pageNum):   
        url='https://www.307ii.com/htm/piclist2/'
        url=url+str(i)+".htm"
        print(url)
        scaleurl=url.replace('/htm/piclist2/',"").replace(str(i)+".htm","")
        print(scaleurl)
        re=s.get(url)
        if re.status_code==200:
            re.encoding='utf-8'
            #print(re.text)
            soup=bs(re.text,'lxml')
            for url in soup.select('div.mainArea ul.textList li a'):
                url1=scaleurl+url['href']
                filename=url.get_text()
                name.append(filename)
                img_url.append(url1)
            return img_url,name

        print('爬虫被禁了',re.status_code)
        return None

def createDir(nm):
    path='chongzi/pic/'+nm+'/'
    if not os.path.exists(path):
        os.makedirs(path)
    return path
    

def crawlpicurl(url,name):
    for (k,nm) in zip(url,name):
        re=s.get(k)
        path=createDir(nm)
        if re.status_code==200:
            re.encoding='utf-8'
            soup=bs(re.text,'lxml')
            count=0
            for tuurl in soup.select('div.mainAreaBlack div.picContent img'):
                count+=1
                src=tuurl['src']
                req=s.get(src)   #对具体图片进行操作了，get操作
                nr=req.content
                filename=path+nm+str(count)+'.jpg'
                print(filename)
                with open(filename,'wb') as f:
                    f.write(nr)        
        else:
            print('爬虫被禁了')   
        print('下载完成',nm)        


if __name__ == "__main__":
    pool=Pool(5)
    url,name=get_img_url()
    
    pool.apply_async(crawlpicurl(url,name))
    pool.close()
    pool.join()

    
```
