```python
# 要解决的问题就是把一列拆分多列
import numpy as np
import pandas as pd
from collections import namedtuple, OrderedDict


df =pd.read_csv('labelstest.csv')

df[["filename","xmin","ymin","xmax","ymax","occluded","class","subclass"]]=df['wocao'].apply(lambda x: pd.Series([i for i in x.split(" ")]))
df=df.drop('wocao',1)  #删除wocao的那一列
df.head()

```

```python
#这个是将class那一列取出来，然后将那一列的数据以list方式循环读取出来，然后使用eval函数
#去掉“”，然后再转化为padas dataframe形式的

cc=[]
for k in df['class']:
    c=eval(k)
    cc.append(c)
#print(cc)

from pandas.core.frame import DataFrame
k={"class1":cc}   #cc是list类型
data=DataFrame(k)
#print(data)

df=pd.concat([df,data],axis=1)   #代表拼接在一起，并且axis=1，代表行对齐
df=df.drop('class',1)
df=df.drop('subclass',1)
df=df.drop('occluded',1)
df.head()
```

```python
df.to_csv('xin1.csv',index=0)
```
