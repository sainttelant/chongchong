```python
import requests
from bs4 import BeautifulSoup as bs
import os
import sys
import time
import random
from io import BytesIO
from PIL import Image
import re

s=requests.Session()
s.keep_alive = False

headers ={
   
    'accept-encoding': 'gzip, deflate, br',
    'accept-language': 'zh-CN,zh;q=0.9',
    'referer': 'https://car.autohome.com.cn/pic/series/4096.html',
    'upgrade-insecure-requests': '1',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36' 
}




if __name__=="__main__":
    print("begin to crawl car pics!")
    url = input("请赋值车型的图片入口：") # https://car.autohome.com.cn/pic/brand-60.html
    path = 'D:/datasets/Carimages/'
    re= s.get(url, headers = headers)
    print("res:",re.status_code)
    soup=bs(re.text,"lxml")
    #print(soup)
    for info in soup.select("div.breadnav a "):
        classname = info.get_text()
        print('classname:',classname)
    pathwhole = path+classname
    if os.path.exists(pathwhole)== False:
        print("创建文件夹%s"%(classname))
        os.makedirs(pathwhole)
    
    import re
    for urllist in soup.select("div.uibox-con.carpic-list02 ul li div span,fn-left a"):
        stringa =str(urllist.find_all('a')[0])
        """
        data = '<a href="/5.mp3" singer="陈慧琳">记事本</a>'
        regex = '<a href=(.*?)singer="(.*?)">(.*?)</a>'
        print(re.match(regex,data,re.DOTALL).groups())
        
        """
        regex = '<a href=(.*?) title=(.*?)</a>'
        zifu = eval(re.match(regex, stringa,re.DOTALL).groups()[0])
        zifunew = zifu.split(".html")[0]+"-1.html"+zifu.split(".html")[1]
        href = 'https://car.autohome.com.cn'+ zifunew
        ret = s.get(href,headers=headers)
        print("ret_astatus:",ret.status_code)
        soupsub=bs(ret.text,'lxml')
        for picurl in soupsub.select("ul li a img"):
            rawpic = "https:"+picurl['src']
            print(rawpic)
            print('rawpic:',rawpic[-5:])
            if "png" in rawpic[-5:]:  
                picname = picurl['src'].split("png")[0].split("/")[-1]+"png"
            elif "jpg" in rawpic[-5:]:
                picname = picurl['src'].split("jpg")[0].split("/")[-1]+"jpg"
            print('picname:',picname)
            #time.sleep(random.uniform(1,3))
            ht=s.get(rawpic)
            if ht.status_code == 200: 
                file = BytesIO(ht.content)
                img = Image.open(file)
                #print('current foldname',foldname)
                try:
                    
                    img.save(pathwhole+"/"+picname)
                except:
                    img.save(pathwhole+"/"+picname+".jpg")
            else:
                print("IP 图片返回有问题，可能被禁网了")
```
