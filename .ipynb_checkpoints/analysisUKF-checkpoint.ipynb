```python
# coding=utf-8
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

tops=[]
lefts=[]
deltat=[]
with open("333Obj.txt","r") as f:
    for line in f.readlines():
        if line[0]=="b":
            top  = float(line.split("\t")[0].split(":")[1])
            left = float(line.split("\t")[1])
            tops.append(top)
            lefts.append(left)
        elif line[0:2]=="de":
            timed = float(line.split("\t")[1])
            deltat.append(timed)
            
            
plt.figure(figsize=(10, 10), dpi=100)
plt.scatter(tops, lefts)
# plt.grid()
# plt.show()   

```

```python
import numpy as np

# 将直角坐标转换为极坐标，可以用坐标中最小的点为极点，也可以用(0,0)为极点
# 半径
def get_radius(x, y):
    r = np.sqrt((x - x_min) ** 2 + (y - y_min) ** 2)
    # r = np.sqrt((x **2+y **2))
    return r


# 角度
def get_angle(x, y):
    a = np.arctan((y - y_min) / (x - x_min)) * 180 / np.pi
    # a = np.arctan(y / x * 180 / np.pi)
    return a

def calculateV(point1,point2,dt):
    num = len(point1)
    v_r=[]
    v_th=[]
    print(num)
    for i in range(0,num-1):
        v_radius = (point1[i+1]-point1[i])/dt[i]
        v_theta = (point2[i+1]-point2[i])/dt[i]
        v_r.append(v_radius)
        v_th.append(v_theta)
        print(v_radius,v_theta)
    return v_r,v_th
x_min = 0
y_min = 0

radius = []
angle = []

for x, y in zip(tops, lefts):
    radius.append(get_radius(x, y))
    angle.append(get_angle(x, y))

radius = np.array(radius)
angle = np.array(angle)
deltat = np.array(deltat)

print(radius,angle)


# 保留小数点后8位
a,b=calculateV(radius,angle,deltat)
```

```python
len(a)
```

```python
len(b)
accerate =[]
for i in range(0,len(a)-1):
    accr=(a[i+1]-a[i])/deltat[i+1]
    acct=(b[i+1]-b[i])/deltat[i+1]
    print("accerate:%.5f,time:%.5f"%(np.sqrt(accr**2+acct**2),deltat[i+1]))
    accerate.append(np.sqrt(accr**2+acct**2))

    
```

```python
sorted_id = sorted(range(len(accerate)), key=lambda k: accerate[k], reverse=True)
sorted_id = [x+1 for x in sorted_id]
print("加速度：",accerate)
print("加速度从大到小排列为：",sorted_id)
print("总加速度的各式为：",len(sorted_id))
415 in sorted_id
```
