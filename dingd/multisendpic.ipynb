```python

import os,re
import subprocess
import time
import sys
import requests
from bs4 import BeautifulSoup as bs



s = requests.session()
s.verify = False


# global param
picname=[]
picurl=[]
pt=[]
fasong=[]
s=requests.Session()
s.keep_alive = False

headers={'Host':'www.mzitu.com',
         'DNT':'1',
         'Upgrade-Insecure-Requests':'1',
         'User-Agent':'Mozilla/5.0 (X11; Linux x86_64; rv:57.0) Gecko/20100101 Firefox/57.0'
        }

headers1 = {
            'Accept': 'text / html, application / xhtml + xml, application / xml;q = 0.9, image / webp, image / apng, * / *;q = 0.8',
            'Accept - Encoding': 'gzip, deflate',
            'Accept - Language': 'zh - CN, zh;q = 0.9',
            'Cache - Control': 'max - age = 0',
            'Connection': 'keep - alive',
            'Host': 'i.meizitu.net ',
            # 'Referer': 'http://www.mzitu.com/139042/7',
            'Referer': 'http://www.mzitu.com/',
            'Upgrade - Insecure - Requests': '1',
            'User - Agent': 'Mozilla / 5.0(Windows NT 10.0; WOW64) AppleWebKit / 537.36(KHTML, like Gecko) Chrome / 66.0.3359.117 Safari / 537.36'
        }
url='http://www.mzitu.com/'

# standard stuff

WORDS = [u"KANMEIZI", u"FAGEMEIZI", u"HAOKANDETU",u"MEINVTU",u"MEIZITU"]
SLUG = "sendpic"

def get_pic_link():
    re=s.get(url,headers=headers)
    re.encoding='utf-8'
    if re.status_code==200:
        soup=bs(re.text,'lxml')
        for pic_link in soup.select('div.postlist li span a'):
            path='mzt/'+pic_link.get_text()+'/'
            pt.append(path)
            if not os.path.exists(path):
                os.makedirs(path)
                picname.append(pic_link.get_text())
                picurl.append(pic_link['href'])
            if os.path.exists(path):
                print('大脑里有这幅图了，直接lvguo')
        return picname,picurl,pt
    else:
        print('IP封了')
        mic.say(u'妹的，姐姐我的Ip好像被封了，亲来帮我修修')
        return None
        
def downpic(url,name,pt):
    for i in range(1,10):
        urln=url+"/"+str(i)
        re=s.get(urln,headers=headers)
        if re.status_code==200:
            #mic.say(u'你个好色之徒，我来帮你搞图片啊，你等着')
            soup=bs(re.text,'lxml')
            for tuurl in soup.select('div.main-image p a img'):
                src=tuurl['src']
                src=src
                req=s.get(src,headers=headers1)
                with open(pt+name+str(i)+'.jpg','wb') as f:
                    f.write(req.content)
                    f.close()
                fasong.append(pt+name+'.jpg')
            print('done')
        else:
            #mic.say(u'不好意思，亲，下载不下来啊图片，你检查一下')
            print('failed')
    return fasong
        
def handle(text, mic, profile, wxbot=None):
    """
        Reports the current time based on the user's timezone.

        Arguments:
        text -- user-input, typically transcribed speech
        mic -- used to interact with the user (for both input and output)
        profile -- contains information related to the user (e.g., phone
                   number)
        wxbot -- wechat bot instance
    """
    sys.path.append(mic.dingdangpath.LIB_PATH)
    from app_utils import sendToUser
    
    dest_path = os.path.expanduser('~/pictures')
    send_to_user = True
    
    # read config
    if profile[SLUG] and 'enable' in profile[SLUG] and        profile[SLUG]['enable']:    
            
        if 'dest_path' in profile[SLUG] and            profile[SLUG]['dest_path'] != '':
            dest_path = profile[SLUG]['dest_path']
 
        if 'send_to_user' in profile[SLUG] and            not profile[SLUG]['send_to_user']:
            send_to_user = False
      
        try:
            if not os.path.exists(dest_path):
                os.makedirs(dest_path)
        except Exception:
            mic.say(u"抱歉，照片目录创建失败")
            return
        
        #url,name=get_img_url()
        na,url,pt=get_pic_link()
        for a,b,c in zip(na,url,pt):  
            fs=downpic(b,a,c)
        
        #下面是全部发送到微信  
        for dest_file in fs:
            if send_to_user:
                target = '邮箱'
                if wxbot is not None and wxbot.my_account != {} and                ('prefers_email' not in profile or
                   not profile['prefers_email']):
                    target = '微信'


                mic.say(u'下载妹子成功！正在发送照片到您的%s' % target)
                if sendToUser(profile, wxbot, u"亲，小黄图收好，太色了你哦", "", [dest_file], []):
                    mic.say(u'发送给你成功')
                else:
                    mic.say(u'发送失败了')
    else:
        mic.say(u"请先在配置文件中开启发送小图片功能")


def isValid(text):
    """
        Returns True if input is related to the time.

        Arguments:
        text -- user-input, typically transcribed speech
    """
    return any(word in text for word in ["看妹子", "发个妹子" ,"好看的图","美女图","妹子图"])

```
